# Basic Matplotlib types with example

```python
# Line Plot:

import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 4, 1, 5, 3]

# Create the plot
plt.plot(x, y)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Line Plot Example")

# Display the plot
plt.show()
```

```python
# Scatter Plot:

import matplotlib.pyplot as plt

# Sample data
x = [1, 3, 5, 7, 2, 4, 5, 6, 1, 2, 1, 3]
y = [4, 6, 2, 8, 1, 4, 5, 7, 8, 4, 5, 6]

# Create the scatter plot with blue circles
plt.scatter(x, y, color='purple', marker='o')

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot Example")

# Display the plot
plt.show()
```

```python
# Bar Plot:

import matplotlib.pyplot as plt

# Categories and their values
categories = ['A', 'B', 'C', 'D']
values = [5, 3, 7, 1]

# Create the bar plot
plt.bar(categories, values)

# Add labels and title
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Bar Plot Example")

# Display the plot
plt.show()
```

```python
# Histogram:

import matplotlib.pyplot as plt
import numpy as np

# Sample data (normally distributed)
data = np.random.normal(loc=5, scale=2, size=1000)

# Create the histogram with 20 bins
plt.hist(data, bins=20)

# Add labels and title
plt.xlabel("Data Values")
plt.ylabel("Frequency")
plt.title("Histogram Example")

# Display the plot
plt.show()
```

```python
# Pie Chart:

import matplotlib.pyplot as plt

# Category labels and their values
categories = ['Apples', 'Bananas', 'Oranges', "Chiku", "Strawberry"]
values = [20, 15, 25, 27, 23]

# Create the pie chart with colors
plt.pie(values, labels=categories, autopct="%1.1f%%", startangle=140, colors=['red', 'yellow', 'green', 'blue', 'purple'])

# Add title
plt.title("Pie Chart Example")

# Display the plot
plt.show()
```

```python
# Box Plot:

import matplotlib.pyplot as plt

# Sample data (multiple datasets)
data1 = [2, 5, 7, 1, 8]
data2 = [4, 1, 9, 3, 6]
data3 = [7, 2, 4, 8, 5]

# Create the box plot with labels
plt.boxplot([data1, data2, data3], labels=['Group 1', 'Group 2', 'Group 3'])

# Add labels and title
plt.xlabel("Groups")
plt.ylabel("Values")
plt.title("Box Plot Example")

# Display the plot
plt.show()
```

```python
# Error Bar Plot:

import matplotlib.pyplot as plt
import numpy as np

# Sample data with errors
x = np.linspace(0, 5, 10)
y = np.sin(x)
yerr = 0.2 * np.random.rand(len(x))  # Simulate random errors

# Create the error bar plot
plt.errorbar(x, y, yerr=yerr, fmt='o-', color='black')  # 'o-' specifies markers and line style

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Error Bar Plot Example")

# Display the plot
plt.show()
```

```python
# Stacked Bar Chart:

import matplotlib.pyplot as plt

# Categories and their values for multiple groups
categories = ['A', 'B', 'C']
values1 = [2, 4, 1]
values2 = [3, 1, 5]

# Create the stacked bar chart
plt.bar(categories, values1, label='Group 1')
plt.bar(categories, values2, bottom=values1, label='Group 2')

# Add labels, title, and legend
plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Stacked Bar Chart Example")
plt.legend()

# Display the plot
plt.show()
```

```python
# Contour Plot:

import matplotlib.pyplot as plt
import numpy as np

# Sample function (2D Gaussian)
def func(x, y):
  return np.exp(-(x**2 + y**2))

# Generate data grid
x = np.linspace(-2, 2, 100)
y = np.linspace(-2, 2, 100)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)

# Create the contour plot with colormap
plt.contour(X, Y, Z, cmap='coolwarm')  # 'coolwarm' is a colormap

# Add labels and title
plt.colorbar(label='Value')  # Colorbar for legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Contour Plot Example")

# Display the plot
plt.show()
```

```python
# 3D Surface Plot:

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Sample function (3D plane)
def func(x, y):
  return x + y

# Generate data grid
x = np.linspace(-2, 2, 20)
y = np.linspace(-2, 2, 20)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create the 3D surface plot with colormap
ax.plot_trisurf(X, Y, Z, cmap='viridis')  # 'viridis' is a colormap

# Add labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.title("3D Surface Plot Example")

# Set viewing angles (optional)
ax.view_init(elev=15, azim=-60)  # Adjust elevation and azimuth for better view

# Display the plot
plt.show()
```

```python
# 3D Wireframe Plot:

#from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Sample function (3D sphere)
def func(x, y):
  return np.sqrt(1 - x**2 - y**2)

# Generate data grid
x = np.linspace(-1, 1, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create the 3D wireframe plot
ax.plot_wireframe(X, Y, Z, rstride=2, cstride=2)  # Control wireframe density

# Add labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.title("3D Wireframe Plot Example")

# Set viewing angles (optional)
ax.view_init(elev=20, azim=-60)  # Adjust elevation and azimuth for better view

# Display the plot
plt.show()
```

```python
# Filled Contour Plot:

import matplotlib.pyplot as plt
import numpy as np

# Sample function (2D parabola)
def func(x, y):
  return x**2 - y

# Generate data grid
x = np.linspace(-2, 2, 40)
y = np.linspace(-2, 2, 40)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)

# Create the filled contour plot with colormap
plt.contourf(X, Y, Z, cmap='plasma')  # 'plasma' is a colormap

# Add labels and title
plt.colorbar(label='Value')  # Colorbar for legend
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Filled Contour Plot Example")

# Display the plot
plt.show()
```

```python
# Ternary Plot:

import matplotlib.pyplot as plt
import matplotlib.tri as tri

# Sample data points (compositions)
data = np.array([[0.3, 0.5, 0.2], [0.6, 0.2, 0.2], [0.1, 0.8, 0.1]])

# Create the triangular grid
fig, ax = plt.subplots()
triang = tri.Triangulation(data[:, 0], data[:, 1])

# Create the ternary plot with labels
ax.scatter(data[:, 0], data[:, 1], color='red')  # Plot data points
ax.set_aspect('equal')  # Equal aspect ratio for triangle
ax.set_xlabel('Component 1')
ax.set_ylabel('Component 2')
ax.set_title('Ternary Plot Example (Component 3 = 1 - (Component 1 + Component 2))')

# Display the plot
plt.show()
```

```python
# Arrow Plot:

import matplotlib.pyplot as plt
import numpy as np

# Sample starting points and directions (arrows)
x_start = np.array([1, 2, 3])
y_start = np.array([2, 1, 4])
x_end = x_start + 0.5  # Arrow head displacement in the x direction
y_end = y_start - 0.3  # Arrow head displacement in the y direction

# Create separate arrows with different styles and colors (using a loop)
for i in range(len(x_start)):
    color = ['red', 'blue', 'green'][i]  # Select color based on index
    plt.arrow(x_start[i], y_start[i], x_end[i] - x_start[i], y_end[i] - y_start[i],
              head_width=0.1, head_length=0.2, color=color, linewidth=2)

# Add labels and title
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Arrow Plot Example")

# Display the plot
plt.show()
```

```python
# Triangulated Mesh Plot (3D Surface with Triangles):

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Sample function (3D plane)
def func(x, y):
  return x + 2*y

# Generate data grid
x = np.linspace(-2, 2, 20)
y = np.linspace(-1, 1, 20)
X, Y = np.meshgrid(x, y)
Z = func(X, Y)

# Create a 3D figure
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create the triangulated mesh plot
triang = tri.Triangulation(X.ravel(), Y.ravel(), Z.ravel())  # Flatten data for triangulation
ax.plot_trisurf(triang, Z, cmap='viridis')  # 'viridis' is a colormap

# Add labels and title
ax.set_xlabel("X-axis")
ax.set_ylabel("Y-axis")
ax.set_zlabel("Z-axis")
plt.title("Triangulated Mesh Plot (3D Surface with Triangles) Example")

# Set viewing angles (optional)
ax.view_init(elev=15, azim=-60)  # Adjust elevation and azimuth for better view

# Display the plot
plt.show()
```

```python
# Scikit-learn Integration:

from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Sample data (2D blobs)
X, y_true = make_blobs(n_samples=300, centers=4, random_state=42)

# Perform K-Means clustering with 4 clusters
kmeans = KMeans(n_clusters=4, random_state=42)
kmeans.fit(X)
y_pred = kmeans.predict(X)

# Create a scatter plot with cluster colors
plt.scatter(X[:, 0], X[:, 1], c=y_pred, cmap='viridis')  # Color by cluster assignment

# Add labels and title
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.title("Scatter Plot with K-Means Clustering Results (using scikit-learn)")

# Display the plot
plt.show()
```

```python
# Interactive Plots:

# This example requires additional libraries like plotly or bokeh
from bokeh.plotting import figure, show
from bokeh.transform import linear_cmap
from bokeh.models import ColumnDataSource

# Sample data (correlated)
x = list(range(11))
y0 = x
y1 = [10 - i for i in x]
y2 = [abs(i - 5) for i in x]

# Create a ColumnDataSource for the data
source = ColumnDataSource(data=dict(x=x, y0=y0, y1=y1, y2=y2))

# Create a line plot with colormapper for interactivity (using bokeh)
mapper = linear_cmap(field_name="y0", palette="Viridis256", low=min(y0 + y1 + y2), high=max(y0 + y1 + y2))
p = figure(width=500, height=250)
p.line(x="x", y="y0", line_width=2, color=mapper, source=source, legend_label="Line 1")
p.line(x="x", y="y1", line_width=2, color=mapper, source=source, legend_label="Line 2")
p.line(x="x", y="y2", line_width=2, color=mapper, source=source, legend_label="Line 3")

show(p)
```
