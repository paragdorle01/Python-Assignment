```python
# 8. Subplots and GridSpec:

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
# Sample data (days, stock prices)
days = np.arange(1, 31)
prices = np.random.randint(100, 150, size=30)

# Calculate moving averages using rolling window with pandas (more efficient)
window_size = 5
maverages = pd.Series(prices).rolling(window=window_size).mean().iloc[window_size - 1:]

# Create a figure with 2 subplots (one for price, one for moving average)
fig, (ax1, ax2) = plt.subplots(2)

# Plot price on the first subplot
ax1.plot(days, prices, label='Stock Price')
ax1.set_xlabel('Days')
ax1.set_ylabel('Price')
ax1.legend()

# Plot moving average on the second subplot with adjusted days
ax2.plot(days[window_size - 1:], maverages, label='Moving Average ({} days)'.format(window_size))
ax2.set_xlabel('Days')
ax2.set_ylabel('Moving Average')
ax2.legend()

# Set title for the overall figure
fig.suptitle('Stock Price and Moving Average')

plt.tight_layout()  # Adjust spacing between subplots
plt.show()
```

```python
# 9. 3D Surface Plots:

from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

# Define a mathematical function (replace with your desired function)
def f(x, y):
  return np.sin(x) * np.cos(y)

# Generate data points
X = np.linspace(-2*np.pi, 2*np.pi, 40)
Y = np.linspace(-2*np.pi, 2*np.pi, 40)
X, Y = np.meshgrid(X, Y)
Z = f(X, Y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the surface
ax.plot_surface(X, Y, Z, cmap='viridis')  # Adjust colormap for different visualizations

# Add labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Surface Plot')

plt.show()
```

```python
# 10. Vector Field Plots:

import matplotlib.pyplot as plt
import numpy as np

# Sample wind data (x, y, wind speed, wind direction)
x = np.arange(0, 10, 0.25)
y = np.arange(0, 10, 0.25)
u = np.sin(2*np.pi*x) * np.cos(2*np.pi*y)  # Sample wind speed values
v = np.cos(2*np.pi*x) * np.sin(2*np.pi*y)  # Sample wind direction values

# Create the vector field plot
plt.quiver(x, y, u, v, color='blue')  # Adjust color and arrow properties

# Add labels and title
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Vector Field Plot (Sample Wind Data)')

plt.show()
```

```python
# 1. FacetGrid and PairGrid:

import seaborn as sns
import pandas as pd

# Sample data
data = pd.DataFrame({'age': [25, 32, 40, 18, 55, 28], 'income': [50000, 70000, 85000, 30000, 90000, 65000], 'category': ['A', 'B', 'A', 'C', 'B', 'A']})

# FacetGrid to see distribution of income across age groups
g = sns.FacetGrid(data, col='category', col_wrap=2)  # Adjust columns for display
g.map(sns.histplot, 'income')  # Create histograms in each subplot

# PairGrid to see relationships between all variables
g = sns.PairGrid(data)
g.map_diag(sns.histplot)  # Histograms on diagonal
g.map_offdiag(sns.scatterplot)  # Scatter plots for off-diagonal elements

plt.show()
```
